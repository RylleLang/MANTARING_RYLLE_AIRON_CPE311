```python
new1 = pd.read_csv('faang.csv', index_col='date', parse_dates=True).assign(
 trading_volume=lambda x: pd.cut(x.volume, bins=3, labels=['low', 'med', 'high'])
 )
new1.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
new1.agg({
    'open': np.mean,
    'high': np.max,
    'low': np.min,
    'close': np.mean,
    'volume': np.sum
 })
```

```python
new1.agg({
    'open': 'mean',
    'high': ['min', 'max'],
    'low': ['min', 'max'],
'close': 'mean'
 })
```

```python
new1.groupby('trading_volume').mean()
```

```python
new1.groupby('trading_volume')['close'].agg(['min', 'max', 'mean'])
```

```python
new_agg = new1.groupby('trading_volume').agg({
 'open': 'mean',
 'high': ['min', 'max'],
 'low': ['min', 'max'],
 'close': 'mean'
 })
new_agg
```

```python
new_agg.columns
```

```python
new_agg.columns = ['_'.join(col_agg) for col_agg in new_agg.columns]
new_agg.head()
```
