```python
import pandas as pd
import numpy as np

LED = '/content/Life Expectancy Data.csv'
LEData = pd.read_csv(LED)
```

# BASED ON THE COLUMNS TYPE THAT HAVE GIVEN IT THINK THAT THE MAIN REASON FOR THIS DATA IS TO LIST THE LIFESTYLE OF A PERSON AND TO KNOW HOW LONG IT LIFE COULD TAKE.

```python
LEData.head()
```

```python
LEData.sort_values(by=['Year'], inplace=True)
LEData
```

```python
LEData.info()
```

```python
LEData.isnull().sum()
```

```python
columns_with_missing_values = LEData.columns[LEData.isnull().any()]

for column in columns_with_missing_values:
    if LEData[column].dtype in ['int64', 'float64']:
        LEData[column].fillna(LEData[column].median(), inplace=True)

LEData['Country'] = LEData['Country'].astype('category')
LEData['Life Expectancy Decade'] = (LEData['Life expectancy '].apply(lambda x: (x // 10) * 10))
```

```python
from sklearn.preprocessing import StandardScaler

numerical_columns = ['Adult Mortality', 'infant deaths', 'Alcohol']
scaler = StandardScaler()

LEData[numerical_columns] = scaler.fit_transform(LEData[numerical_columns])
```

# FOR THE VISUAL DATA

```python
x = LEData['Schooling']
y = LEData['Life expectancy ']
```

```python
LEData.dropna(subset=['Schooling', 'Life expectancy '], inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x_reshaped = x.values.reshape(-1, 1)

linear_regression = LinearRegression()
linear_regression.fit(x_reshaped, y)

y_pred = linear_regression.predict(x_reshaped)
```

# THIS ARE THE STUDENTS SCATTER PLOT IN YEARS

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, color='blue', label='Data Points')

plt.plot(x, y_pred, color='red', label='Regression Line')

plt.xlabel("Schooling (years)")
plt.ylabel("Life Expectancy (years)")
plt.title("Scatter Plot with Simple Linear Regression")
plt.legend()
plt.show()
```

# ALSO IN LINE CHART FOR MUST EASIER TO VIEW THE VALUES

```python
LEData.head()
LEData['Year'].unique()
LEData.isnull().sum()
```

```python
LEData_grouped = LEData.groupby('Year')['Life expectancy '].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Life expectancy ', data=LEData_grouped, marker='o', linestyle='-', color='b')

plt.xlabel("Year")
plt.ylabel("Average Life Expectancy")
plt.title("Trend of Life Expectancy Over Time")
plt.show()
```
