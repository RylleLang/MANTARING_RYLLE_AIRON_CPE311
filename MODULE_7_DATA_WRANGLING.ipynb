# Exercise 3


*   Using web scraping, search for the list of the hospitals, their address and contact information. Save the list in a new csv file, hospitals.csv.
*   Using the generated hospitals.csv, convert the csv file into pandas dataframe. Prepare the data using the necessary preprocessing techniques.



```python
import requests
from bs4 import BeautifulSoup
import csv


url = "https://example.com/hospitals"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

hospital_elements = soup.find_all("div", class_="hospital-info")

hospitals = []
addresses = []
contacts = []

for hospital in hospital_elements:
    hospital_name = hospital.find("h2").text.strip()
    address = hospital.find("p", class_="address").text.strip()
    contact_info = hospital.find("p", class_="contact").text.strip()

    hospitals.append(hospital_name)
    addresses.append(address)
    contacts.append(contact_info)

with open("hospitals.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Hospital", "Address", "Contact"])
    for data in zip(hospitals, addresses, contacts):
        writer.writerow(data)

print("Data has been scraped and saved to hospitals.csv")
```
