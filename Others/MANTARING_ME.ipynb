```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
census_income = fetch_ucirepo(id=20)

# data (as pandas dataframes)
X = census_income.data.features
y = census_income.data.targets

# metadata
print(census_income.metadata)

# variable information
print(census_income.variables)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data=X, columns=census_income.data.feature_names)
df['income'] = y  # Assuming 'income' is the target

# Check for missing values
print(df.isnull().sum())
```

```python
#visualize the distribution of the target
sns.countplot(x='income', data=df)
plt.title('Income Distribution')
plt.show()
```

In summary it show that more people only earn in lower than 50k

```python
#visualize the relationship between age and income
sns.boxplot(x='income', y='age', data=df)
plt.title('Income by Age')
plt.show()
```

In this plot it show that the age 40 above or the older ones is getting a high amout of income than the 40 below

```python
# For instance, relationship between education level and income
plt.figure(figsize=(10, 6))
sns.countplot(x='education', hue='income', data=df)
plt.title('Income by Education Level')
plt.xticks(rotation=45)
plt.show()
```

In the plot it show that High school grad is had less income and more population of it rather than a college grad that earn greater than 50k

```python
#visualize the relationship between age and hours-per-week
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='hours-per-week', data=df, hue='income', alpha=0.5)
plt.title('Scatter Plot: Age vs Hours-per-week')
plt.xlabel('Age')
plt.ylabel('Hours-per-week')
plt.legend(title='Income', loc='upper right')
plt.show()
```

We get a plot that show that the less than or equal to 50k of all of age is more

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, color='red', linewidth=2, linestyle='-', label='Age')
sns.histplot(df['education-num'], kde=True, color='orange', linewidth=2, linestyle='--', label='Education Num')
plt.title('Line Histogram: Distribution of Age and Education Num')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

In the graph show that the educational attainment is having more impact rather than the age of a people

```python
numerical_cols = df.select_dtypes(include='number').columns

plt.figure(figsize=(12, 8))
for col in numerical_cols:
    plt.subplot(2, 3, list(numerical_cols).index(col) + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
