```python
import random

random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):



*   Mean
*   Median
*   Mode (hint: check out the Counter in the collections module of the standard library at
https://docs.python.org/3/library/collections.html#collections.Counter)
*   Sample variance

```python
mean = sum(salaries) / len(salaries) # simply get the sum and divide by the legnth of the value
print("Mean Salary:", mean)
```

```python
sort = sorted(salaries) #sort the salaries
n = len(sort) #legnth of the sorted

if n % 2 == 0:

    median = (sort[n//2 - 1] + sort[n//2]) / 2
else:
    median = sort[n//2]

print("Median Salary:", median)
```

```python
import statistics

mode = statistics.mode(salaries)
print("Mode Salary:", mode)
```

```python
legnth = len(salaries) #legnth of the salaries

mean = sum(salaries) / legnth #to get the mean

sqrd = [(x - mean)**2 for x in salaries] #to get median

variance = sum(sqrd) / (legnth - 1) # to get the varience

print("Sample Variance:", variance)
```

```python
import math

legnth = len(salaries) #legnth of the salaries

mean = sum(salaries) / legnth #to get the mean

sqrd = [(x - mean)**2 for x in salaries] #to get median

variance = sum(sqrd) / (legnth - 1) # to get the varience

deviation = math.sqrt(variance) # to get the standard deviation

print("Sample Standard Deviation:", deviation)
```

# Exercise 2

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:



*   Range
*   Coefficient of variation Interquartile range
*   Quartile coefficient of dispersion



```python
salary = max(salaries) - min(salaries) # simply subtract the min to max of the salaries
print("Range of Salaries:", salary)
```

```python
import statistics

# Calculate the coefficient of variation
mean = statistics.mean(salaries)
deviation = statistics.stdev(salaries)
coefficient = (deviation / mean) * 100  # Multiply by 100 for percentage

print("Coefficient of Variation:", coefficient)

# Calculate the range
r1 = statistics.quantiles(salaries, n=4)[0]
r2 = statistics.quantiles(salaries, n=4)[2]
range = r2 - r1

print("Range:", range)
```

```python
import statistics

# Calculate the interquartile range
r1 = statistics.quantiles(salaries, n=4)[0]
r2 = statistics.quantiles(salaries, n=4)[2]
interquartile_range = r2 - r1

# Calculate the quartile coefficient of dispersion
qcod = (interquartile_range / (r1 + r2)) * 100  # Multiply by 100 for percentage

print("Quartile Coefficient of Dispersion:", qcod)
```
