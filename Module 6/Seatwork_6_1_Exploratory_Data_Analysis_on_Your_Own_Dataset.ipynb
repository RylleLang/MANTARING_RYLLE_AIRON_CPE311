# Female

```python
female = crd[crd['Gender']== 'F'] # to get all females in the data
female
```

```python
meanf = sum(female.Amount) / len(female.Amount) # Get the mean of the amounts in credit cards for only females
print("credit cards of females Mean:", meanf)
```

```python
female.Amount.mean() # this also
```

```python
sort = sorted(female.Amount) # sort the amounts
n = len(sort) #legnth of the sorted

if n % 2 == 0:

    median = (sort[n//2 - 1] + sort[n//2]) / 2
else:
    median = sort[n//2]

print("Median Amounts of Female:", median)
```

```python
female.Amount.mode()
```

```python
female.Amount.sum() # total amount of all females in the credits
```

```python
goldfemale = female[female['Card Type']== 'Gold'] # Gold Users for females
goldfemale
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
goldfemale.groupby('Exp Type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
male = crd[crd['Gender']== 'M']
male
```

```python
meanm = sum(male.Amount) / len(male.Amount) # Get the mean of the amounts in credit cards for only males
print("credit cards of male Mean:", meanm)
```

```python
male.Amount.mean() # or this one
```

```python
sort = sorted(male.Amount) # sort the amounts
n = len(sort) #legnth of the sorted

if n % 2 == 0:

    median = (sort[n//2 - 1] + sort[n//2]) / 2
else:
    median = sort[n//2]

print("Median Amounts:", median)
```

```python
Goldusers = male[male['Card Type']== 'Gold'] # to get the gold card holders of males
Goldusers
```
